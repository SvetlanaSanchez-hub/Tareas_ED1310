```python
import datetime
class users: 
    def __init__(self,numeroTrabajador ,nombre, horasExtras,sueldobase,anioIngreso): 
        self.numeroTrabajador = numeroTrabajador
        self.nombre = nombre 
        self.horasExtras = horasExtras
        self.sueldobase = sueldobase
        self.anioIngreso = anioIngreso
 #creacion de lista
list = []

def readFile():
    if(len(list) <= 0):
        with open('/content/junio.dat', 'r') as f:
            d = f.readlines()
            d.pop(0) #eliminar encabezados
            for i in d:
                k = i.rstrip().split(",")      
                list.append(users(int(k[0]),k[1] + " " + k[2] + " " + k[3],int(k[4]),int(k[5]),int(k[6])))

def HorasExtrasTrabajador():
   for item in list:
    costoHoraExtra = 276.50
    constoHoraUser = item.horasExtras * costoHoraExtra 
    print("El costo de horas extras del trabajador: " + item.nombre + " es de: $" + str(constoHoraUser))
 

def PrestacionAntiguedad():
    anioActual = int(datetime.datetime.now().date().strftime("%Y"))
    for item in list:
        aniosUser = anioActual - item.anioIngreso
        if(aniosUser > 0):
            sueldoXAnioUser = ((aniosUser * 0.03) * 100) + item.sueldobase
            print("El trabajador: " + item.nombre + " tiene " + str(aniosUser) + " años de antiguedad, por contrato su sueldo es:$ " + str(sueldoXAnioUser))
            
def MayorAntiguedad():
        userMayorAntiguedad = users(0, "", 0, 0, 0)
        for row in list:
            if(userMayorAntiguedad.numeroTrabajador == 0):
                userMayorAntiguedad = row
            elif (userMayorAntiguedad.anioIngreso > row.anioIngreso):
                userMayorAntiguedad = row
                print("El empleado con mayor antiguedad es: " + userMayorAntiguedad.nombre)
                
def MenorAntiguedad():
        userMenorAntiguedad = users(0, "", 0, 0, 0)
        for row in list:
            if(userMenorAntiguedad.numeroTrabajador == 0):
                userMenorAntiguedad = row
            elif (userMenorAntiguedad.anioIngreso < row.anioIngreso):
                userMenorAntiguedad = row
                print("El empleado con menor antiguedad es: " + userMenorAntiguedad.nombre)
        
        
        
        

def SalirPrograma():
    return 'Opcion para salir del programa'


def llama_la(opcion='f'):
    # Obtiene la funcion del diccionario
    func = opciones[opcion]['Funcion']
##    print(func)
##    Enter_continuar()
    # Ejecuta la function
    return func()

def cls():
    import os
    _=os.system("cls")# en windows
##    _=os.system("clear") # en unix linux
##    print ("\n" * 25) #pseudo limpiar la pantalla sin os solo la desplaza

def menu(hecho='',s=25):
    readFile()
    print('Estas son las opciones disponibles del programa:','\n','-'*s)
    for op in opciones:
        print('\t',op,' : ',opciones[op]['Des'])
    #print('-'*s,'Ya realice->',hecho)
    op=input('Opcion--->').lower()
    if op in opciones:
        print(llama_la(opcion=op))
        if op=='f':
            return op
    else:
        print('Opcion inexistente---',op,'---intente de nuevo')
        return ''
    return op

opciones ={
'1' : {'Des': 'Obtiene valor de cada trabajador de sus horas extras', 'Funcion': HorasExtrasTrabajador},
'2' : {'Des': 'Contracto colectivo', 'Funcion': PrestacionAntiguedad},
'3' : {'Des': 'Mayor antiguedad', 'Funcion': MayorAntiguedad},
'4' : {'Des': 'Menor antiguedad', 'Funcion': MenorAntiguedad},
'f' : {'Des': 'Terminar', 'Funcion': SalirPrograma}
}


op,hecho,s='','',25
while op!='f':
    op=menu(hecho,s)
    hecho=hecho +' ' + op
    cls()
print('-'*s,'Fin del programa')

```
